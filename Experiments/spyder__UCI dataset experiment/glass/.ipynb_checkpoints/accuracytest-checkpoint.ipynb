```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('APCOS_0_pred.txt');
data1.head()
```

```python
data1.groupby(by=['Actual','Predicted']).size()
```

```python
noclasses=np.unique(data1.Predicted).shape[0]
noclasses
```

```python
temp=data1.copy()
```

```python
temp
```

```python
from itertools import permutations
combcols=(list(permutations(np.arange(0,noclasses,1), 6)))
len(combcols)
```

```python
combcols[10]
```

```python
temp.Predicted[2]
```

```python
currentlist=combcols[20]
for i in range(temp.shape[0]):
    temp.iloc[i,1] = currentlist[temp.Predicted[i]]
    print(temp.Predicted[i],temp.iloc[i,1],i)
```

```python
# temp
```

```python
from sklearn import metrics as m
```

```python
accuracylist=[]
for j in range(len(combcols)):
    currentlist=combcols[j]
    temp=data1.copy()
    for i in range(temp.shape[0]):
        temp.iloc[i,1] = currentlist[temp.Predicted[i]]
    accuracylist.append(m.accuracy_score(temp.Actual,temp.Predicted))
    print(j)

currentlist=combcols[np.argmax(accuracylist)]
temp=data1.copy()
for i in range(temp.shape[0]):
    temp.iloc[i,1] = currentlist[temp.Predicted[i]]
    
prf=m.precision_recall_fscore_support(temp.Actual,temp.Predicted,average='weighted')
print('Accuracy = ',accuracylist[np.argmax(accuracylist)]) 
print('Precision = ',prf[0])
print('Recall = ',prf[1])
print('F1_score = ',prf[2])
```
