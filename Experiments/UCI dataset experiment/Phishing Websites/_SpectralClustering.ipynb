```python
cluster=2
# data=pd.read_csv('raw_pe_images.csv', header=0, sep=',')  #
data=pd.read_csv('kr-vs-kp.data', header=None, sep=',')
datasetname='iris'
x= np.asarray(data.iloc[:, :-1]) 
y= np.asarray(data.iloc[:, -1]) 
le=LabelEncoder()
y=le.fit_transform(y) 
data.tail() 
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()

for i in range(x.shape[1]):
    x[:,i]=ord_enc.fit_transform(x[:,i:i+1]).ravel()
cluster=np.unique(y).shape[0]
```

```python
for i in range(10):
    
    data=shuffle(data)
    x= np.asarray(data.iloc[:, 1:-1]) 
    y= np.asarray(data.iloc[:, -1]) 
    le=LabelEncoder()
    y=le.fit_transform(y)
    
    for j in range(x.shape[1]):
        x[:,j]=ord_enc.fit_transform(x[:,j:j+1]).ravel()
    

    clustering = SpectralClustering(n_clusters=cluster, affinity='nearest_neighbors' ,
                                    gamma=1,  assign_labels="discretize", random_state=0)
    clustering.fit(x) 
    ny=clustering.labels_   
        
    mypred=pd.DataFrame()
    mypred['Actual']=y
    mypred['Predicted']=ny
    
    mypred.to_csv('SpectralClustering_'+str(i)+'_pred.txt',sep=',',index=False)
    print(i)
    
```
