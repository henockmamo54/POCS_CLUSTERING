```python
cluster=3
# data=pd.read_csv('raw_pe_images.csv', header=0, sep=',')  #
data=pd.read_csv('csv_result-Training Dataset.csv', header=0, sep=',')
datasetname='iris'
x= np.asarray(data.iloc[:, 1:-1]) 
y= np.asarray(data.iloc[:, -1]) 
le=LabelEncoder()
y=le.fit_transform(y) 
cluster=np.unique(y).shape[0]
data.tail() 
```

```python
for i in range(10):
    
    data=shuffle(data)
    x= np.asarray(data.iloc[:, 1:-1]) 
    y= np.asarray(data.iloc[:, -1]) 
    le=LabelEncoder()
    y=le.fit_transform(y) 
    
    clustering = AgglomerativeClustering(n_clusters = cluster,linkage= "single")  
    clustering.fit(x) 
    
    ny=clustering.labels_  
        
    mypred=pd.DataFrame()
    mypred['Actual']=y
    mypred['Predicted']=ny
    
    mypred.to_csv('AgglomerativeClustering_'+str(i)+'_pred.txt',sep=',',index=False)
    print(i)
    
```
