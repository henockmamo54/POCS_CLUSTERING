```python
cluster=8
data=pd.read_csv('pd_speech_features.csv', header=0, sep=',')  
datasetname='ecoli'
x= np.asarray(data.iloc[:, 1:-1]) 
y= np.asarray(data.iloc[:, -1]) 
le=LabelEncoder()
y=le.fit_transform(y)
cluster=np.unique(y).shape[0] 
data.head() 
```

```python
cluster=np.unique(y).shape[0]
cluster
```

```python
for i in range(10):
    
    data=shuffle(data)
    x= np.asarray(data.iloc[:, 1:-1]) 
    y= np.asarray(data.iloc[:, -1]) 
    le=LabelEncoder()
    y=le.fit_transform(y)
    
    kmeans = KMeans(n_clusters = cluster,max_iter=50, random_state=5) 
    kmeans.fit(x) 
    
    ny=kmeans.labels_   
        
    mypred=pd.DataFrame()
    mypred['Actual']=y
    mypred['Predicted']=ny
    
    mypred.to_csv('KMEANS_'+str(i)+'_pred.txt',sep=',',index=False)
    print(i)
    
```
