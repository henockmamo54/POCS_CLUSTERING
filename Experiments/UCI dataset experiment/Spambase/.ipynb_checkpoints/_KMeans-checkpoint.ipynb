```python
import numpy as np
import random
import math
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
import time
import pandas as pd
from scipy.spatial.distance import cdist
from sklearn.utils import shuffle
# from sklearn.cluster import KMeans
import numpy as np
from sklearn.cluster import KMeans 

from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
np.random.seed(0)
```

KMeans - Clustering
===============

```python
cluster=3
# data=pd.read_csv('raw_pe_images.csv', header=0, sep=',')  #
data=pd.read_csv('spambase.data', header=None, sep=',')
datasetname='iris'
x= np.asarray(data.iloc[:, :-1]) 
y= np.asarray(data.iloc[:, -1]) 
le=LabelEncoder()
y=le.fit_transform(y)
cluster=np.unique(y).shape[0]
print(cluster,data.shape)
data.tail() 
```

```python
for i in range(10):
    
    data=shuffle(data)
    x= np.asarray(data.iloc[:, :-1]) 
    y= np.asarray(data.iloc[:, -1])  
    le=LabelEncoder()
    y=le.fit_transform(y) 

    kmeans = KMeans(n_clusters = cluster,max_iter=25, init="random", random_state=5)  
    kmeans.fit(x) 
    
    ny=kmeans.labels_   
        
    mypred=pd.DataFrame()
    mypred['Actual']=y
    mypred['Predicted']=ny
    
    mypred.to_csv('KMEANS_'+str(i)+'_pred.txt',sep=',',index=False)
    print(i)
```
